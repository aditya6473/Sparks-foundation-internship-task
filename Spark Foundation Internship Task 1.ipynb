```python
#Import all the libraries which are required
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
#Load the data which is need to be analysed
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
#Get the first 10 rows of the data
s_data.head(10)
```

```python
s_data.plot(x='Hours', y='Scores', style='+',color='black')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
print("Training complete.")
```

```python
line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y,color='red')
plt.plot(X, line);
plt.show()
```

```python
print(X_test)
y_pred = regressor.predict(X_test)
```

```python
X_test1 = X_test.flatten() # for converting array in one dimension
df = pd.DataFrame({'Hours':X_test1,'Actual score': y_test, 'Predicted Score': y_pred})  
df.head()
```

```python
score_pred = regressor.predict([[9.25]])
print("No of Hours = 9.25")
print("Predicted Score = {}".format(score_pred[0]))
```

```python
plt.title('Hours vs Percentage')
plt.xlabel('Hours')
plt.ylabel('Percentage')
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
